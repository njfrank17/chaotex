# From the sky to Instagram  

*The process required to turn a skydiving experience into a shareable artifact*  

**experience -> compiled_to -> artifact**  

```python
def plan_jump ():
    print("1. PLAN JUMP:")
    print("  a. Assemble jumpers")
    print("  b. Plan jump: dive and safety")
    print("  c. Determine if there is a camera")
    print("")
    return

def obtain_video_from_jumper ():
    print("2. OBTAIN VIDEO FROM JUMPER(S):")
    print("  a. Transfer from GoPro to iPhone (app)")
    print("  b. Save video to iPhone (local storage)")
    print("")
    return

def backup_video_to_drive ():
    print("3. BACKUP VIDEO TO DRIVE:")
    print("  a. Save video to Google Drive:")
    print("    i.  Save to `_content/JumpVideos/`")
    print("    ii. Name as `jump-xxx_short-description-of-jump_mm-dd-yyyy`")
    print("")
    return

def edit_jump_video ():
    print("4. EDIT JUMP VIDEO:")
    print("  a. Download video to laptop")
    print("  b. Use CapCut desktop version to clip, trim, and scale")
    print("  c. Export video from CapCut to `_content/Drafts/`")
    print("    i. Name as `jump-xxx_short-description-of-jump_mm-dd-yyyy`")
    print("")
    return

def edit_draft ():
    print("5. EDIT DRAFT:")
    print("  a. Download draft from Drive: `_content/Drafts/`")
    print("  b. Add to Instagram and edit:")
    print("    i. add captions")
    print("    ii. add audio_track and mix audios")
    print("  c. Caption and tag video")
    print("")
    return

def publish_content ():
    print("6. PUBLISH CONTENT:")
    print("  a. Hit the publish button")
    print("  b. Monitor post and engage")
    return
```

```python
def post_a_jump_video ():  
    plan_jump()
    obtain_video_from_jumper()
    backup_video_to_drive()
    edit_jump_video()
    edit_draft()
    publish_content()
    return
```

```python
post_a_jump_video()
```
